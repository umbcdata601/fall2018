```python
all_import_and_from=[]
for this_notebook in list_of_notebook_files:
    try:
        data = json.load(open(this_notebook,'r'))
    except ValueError:
        print('invalid JSON:',this_notebook)
    for cell_indx in range(len(data['cells'])):
        if (data['cells'][cell_indx]['cell_type']=='code'):
            list_of_lines = data['cells'][cell_indx]['source']
            for this_line in list_of_lines:
                if (this_line.startswith("import")):
                    #print(this_line)
                    all_import_and_from.append(this_line.rstrip())
                if (this_line.startswith("from")):
                    #print(this_line)
                    all_import_and_from.append(this_line.rstrip())
```

```python
uniq_import_and_from = list(set(all_import_and_from))
```

```python
module_list=[]
for this_entry in uniq_import_and_from:
    entry_as_list = this_entry.split()
    if this_entry.startswith('from'):
        module_list.append(entry_as_list[1])
    elif this_entry.startswith('import'):
        module_list.append(entry_as_list[1])
    else:
        print('ERROR FOUND:',this_entry)
```

```python
module_list = list(set(module_list))
module_list.sort() # in-place
module_list
```
