```python
import pandas
pandas.__version__
```

```python
 d = {'static': [156343, 23425344, 59482], 
      'name': ['Bob\'s Best Beds' , 'Gil\'s Great Glasses', 'Pineapple 4 Cheap']}
df1 = pandas.DataFrame(data=d)
df1.to_csv('lookup.csv')
df1
```

```python
 b = {'static': [156343, 23425344, 59482,156343,59482],
      'experiment_index':[593,484,493,234,987], # unique
      'color':['black','blue','red','green','gray']}
df2 = pandas.DataFrame(data=b)
df2.to_csv('color.csv')
df2
```

```python
 c = {'experiment_index':[593,484,493,234,987,493,593],
      'count':[3,  4,  1,  2,  8,  9,  2],
      'smell':['good','bad','bad','good','bad','good','good']}
df3 = pandas.DataFrame(data=c)
df3.to_csv('smell.csv')
df3
```
