https://stackoverflow.com/questions/18574108/how-do-convert-a-pandas-dataframe-to-xml

```python
def xml_producer(row):
    xml = ['<item>']
    xml.append('  <email>'+row['email']+'</email>')
    xml.append('  <balance>'+row['balance']+'</balance>')
    xml.append('</item>')
    return '\n'.join(xml)
```

```python
with open('list.xml','w') as fxml:
    fxml.write('<?xml version="1.0" encoding="UTF-8"?>\n')
    fxml.write('<list>\n')
    fxml.write('\n'.join(df.apply(xml_producer, axis=1)))
    fxml.write('\n</list>')
```
