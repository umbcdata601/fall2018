## linear regression using 2 features

http://scikit-learn.org/stable/modules/generated/sklearn.linear_model.LinearRegression.html

```python
from sklearn.linear_model import LinearRegression
```

```python
X = df[["Desertion","Literacy"]]
y = df["Area"]
```

Claim: desertion rate and literacy can be used to predict the area of a region of France

_To revisit later_: how would we test the validity of this claim?

```python
model = LinearRegression().fit(X,y)
```

```python
model.coef_
```

```python
coeff_df = pandas.DataFrame(model.coef_, X.columns, columns=['Coefficient'])  
coeff_df  
```

```python
model.intercept_
```

### plot the plane

```python
import numpy as np

np.linspace(0,1,10)
```

```python
# https://stackoverflow.com/questions/48335279/given-general-3d-plane-equation-how-can-i-plot-this-in-python-matplotlib

des = np.linspace(0,90,10)
lit = np.linspace(0,90,10)

D,L = np.meshgrid(des,lit)
A = model.intercept_ + model.coef_[0]*D + model.coef_[1]*L

fig = plt.figure()
ax = fig.gca(projection='3d')

surf = ax.plot_surface(D,L,A)

_=ax.set_xlabel('Desertion')
_=ax.set_ylabel('Literacy')
_=ax.set_zlabel('Area')
```

## plot plane and data points

```python
ax = plt.figure().gca(projection='3d')
ax.scatter(df["Desertion"], df['Literacy'], df['Area'])

surf = ax.plot_surface(D,L,A)

_=ax.set_xlabel('Desertion')
_=ax.set_ylabel('Literacy')
_=ax.set_zlabel('Area')

plt.show()
```
