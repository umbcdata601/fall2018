```python
import pandas
import sqlite3
pandas.__version__
```

```python
conn = sqlite3.connect("History")
hist_df = pandas.read_sql_query("SELECT datetime(last_visit_time/1000000-11644473600,'unixepoch','localtime'), url FROM urls ORDER BY last_visit_time DESC;", conn)
hist_df.columns=['date_time','url']
hist_df.head()
```

```python
hist_df['date_time'] = pandas.to_datetime(hist_df['date_time'],
                                          format='%Y-%m-%d %H:%M:%S',
                                          errors='coerce')
#,infer_datetime_format=True)
#)
```

Total time span of my browser history:

```python
total_time=hist_df["date_time"][0]-hist_df["date_time"][hist_df.shape[0]-1]
total_time
```

```python
print(hist_df.shape[0],'webpages visited in',total_time.days,'days')
```

Naive analysis is to get the average:

```python
print(hist_df.shape[0]/total_time.days,'webpages visited per day')
```

```python
prev_time=None
this_window_start=None
this_window_end=None
window_durations={}
window_indx=0
for this_time in hist_df["date_time"]:
    if (prev_time is None): # at the first timestamp
        prev_time=this_time
        window_indx+=1
        this_window_start=this_time
    else: # all timestamps after the first
        gap_time=prev_time-this_time
        print('gap:',gap_time,
              ';window',window_indx,
              'duration:',this_window_start-this_time)
#        if (gap_time>pandas.Timedelta(hours=1)): # need to reset window
        if (gap_time>pandas.Timedelta(days=1)): # need to reset window
            # because we don't want to include gap in window, 
            # use prev_time to determine window duration
            window_durations[window_indx]=this_window_start-prev_time
            window_indx+=1
            this_window_start=this_time
        prev_time=this_time
```

```python
window_sum=pandas.Timedelta(hours=0)
for windx,win_delt in window_durations.items():
    window_sum+=win_delt

print(window_sum)
```

### by the hour rather than per day

```python
prev_time=None
this_window_start=None
this_window_end=None
window_durations={}
window_indx=0
for this_time in hist_df["date_time"]:
    if (prev_time is None): # at the first timestamp
        prev_time=this_time
        window_indx+=1
        this_window_start=this_time
    else: # all timestamps after the first
        gap_time=prev_time-this_time
#        print('gap:',gap_time,
#              ';window',window_indx,
#              'duration:',this_window_start-this_time)
        if (gap_time>pandas.Timedelta(hours=1)): # need to reset window
#        if (gap_time>pandas.Timedelta(days=1)): # need to reset window
            # because we don't want to include gap in window, 
            # use prev_time to determine window duration
            window_durations[window_indx]=this_window_start-prev_time
            window_indx+=1
            this_window_start=this_time
        prev_time=this_time
```

```python
window_sum=pandas.Timedelta(hours=0)
for windx,win_delt in window_durations.items():
    window_sum+=win_delt

print(window_sum)
```

```python
window_sum/pandas.Timedelta(days=1)
```

```python
window_sum/total_time
```

Utilization of 0.7% (1 hour window) is in comparison to the 12% (1 day window)

```python
print(hist_df.shape[0]/(window_sum/pandas.Timedelta(hours=1)),'webpages visited per hour')
```
