```python
from faker import Faker
fake = Faker()
```

```python
import time
import pandas
```

```python
def create_df(num_rows):
    df = pandas.DataFrame()
    for i in range(num_rows):
        row_dict={'word':fake.words(1)[0], 
              'state':fake.state(), 
              'num':i/10,
              'city':fake.city(), 
              'another word':fake.words(1)[0],
              'email':fake.email(),
              'name':fake.name(),
              'fixed':fake.url(),
              'val':fake.latitude(),
              'country':fake.country()}
        df = df.append(row_dict, ignore_index=True)
    return df
```

```python
create_df(4)
df
```

it takes 10 seconds per 1000 rows

therefore 4,000,000 would take 11 hours

```python
start = time.time()
df = create_df(1000)
print(time.time() - start,'seconds')
#df
```

1000 rows x 10 gets 1024000 rows which is 133 MB and writes in 20 seconds

1000 rows x 11 gets 2048000 rows which is 269 MB and writes in 40 seconds

1000 rows x 12 gets 4096000 rows which is 540 MB and writes in 73 seconds
