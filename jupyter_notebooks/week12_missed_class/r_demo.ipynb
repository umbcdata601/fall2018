`runif` generates random numbers between 0 and 1. The first argument specifies how many numbers you want.

```r
runif(4)
```

Create a list of 1000 values between 0 and 1

Then scale up to values between 0 and 10

Then round to integers

```r
n <- floor(runif(1000)*10)
n[0:20]
```

`summary` provides statistics about the list

```r
summary(n)
```

the `table` function yeilds the count of each entry in the list

```r
t <- table(n)
t
```

http://www.rexamples.com/6/Plotting%20a%20uniform%20distribution

```r
barplot(t,main='histogram of random values', xlab="Generated Numbers", ylab="Occurrences")
```
