```python
names_df = df[0]
names_df.columns = list(names_df.loc[0]) # use row 0 as the column names
names_df.drop(0).drop('Rank',axis=1)     # drop the "rank" variable since it's redundant with index
```

```python
def generate_random_name():
    if (random.randint(0,1)==0):
        return random.choice(names_df['Male name'])
    else:
        return random.choice(names_df['Female name'])
```

```python
def create_address():
    address={}
    street_number=str(random.randint(9,100))
    street=random.choice(list_of_street_names)
    address["street addr"]=street_number+" "+street
    address["state"]=random.choice(list_of_state_abbr)
    address["apt number"]="apt "+str(random.randint(9,100))
    address["zip code"]=str(random.randint(9999,100000))
    return address
```

```python
def generate_random_phone_number(delim):
    city_code_prefix=str(random.randint(99,1000))
    local_code_prefix=str(random.randint(999,10000))
    return city_code_prefix+delim+local_code_prefix
```

```python
with open('ragged.csv','w') as rcsv:
    rcsv.write("name,phone,address\n")
    for this_row in range(100000):
        csv_delim=","
        address_dict=create_address()
        name=generate_random_name()
        phone=generate_random_phone_number(random.choice(["-"," ","."]))
        dice=random.randint(0,3) # 0,1,2,3
        if (dice==1):
            rcsv.write(name+csv_delim+
                  phone+csv_delim+
                  address_dict['state']+"\n")
        elif (dice==2):
            rcsv.write(name+csv_delim+
                  phone+csv_delim+
                  address_dict['street addr']+csv_delim+
                  address_dict['state']+csv_delim+
                  address_dict['zip code']+"\n")
        elif (dice==3):
            rcsv.write(name+csv_delim+
                  phone+csv_delim+
                  address_dict['state']+csv_delim+
                  address_dict['zip code']+"\n")
        else:
            rcsv.write(name+csv_delim+
                  phone+csv_delim+
                  address_dict['apt number']+csv_delim+              
                  address_dict['street addr']+csv_delim+
                  address_dict['state']+csv_delim+
                  address_dict['zip code']+"\n")
```
